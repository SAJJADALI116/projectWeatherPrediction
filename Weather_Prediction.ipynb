```python
# Importing Libraries
import numpy as np
import pandas as pd
```

```python
# Reading Data from given CSV File
data = pd.read_csv("/Weather.csv")
data
```

```python
# Clearing out unnecessary columns from CSV File
data = data.drop(columns=["date_time", "precipMM", "moonset", "moonrise", "sunset", "sunrise"])
data
```

```python
# Defining a function that will work as classifier
def separator(temperature):
  if temperature < 10:
    return "Very Cold"
  elif temperature < 30:
    return "Normal"
  else:
    return "Very Hot"
```

```python
# Applying Separator
data['maxtempC'] = data['maxtempC'].apply(separator)
data
```

```python
# Storing Max Temperature in y
y = data["maxtempC"]
y
```

```python
# Storing rest of data in x
x = data.drop("maxtempC",axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and test sets, stratifying the target variable
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
# Train the model on the training data
model.fit(xTrain, yTrain)
```

```python
yPrediction = model.predict(xTest)
```

```python
from sklearn.metrics import classification_report

# Print the evaluation report
print(classification_report(yTest, yPrediction))
```

```python
from sklearn.metrics import accuracy_score
# Compute the accuracy of the predictions
accuracy = accuracy_score(yTest, yPrediction)
# Print the accuracy
print(accuracy)
```
